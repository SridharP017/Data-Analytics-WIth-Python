```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Set the base URL and the URL parameters
base_url = "https://www.magicbricks.com/independent-house-for-sale-in-coimbatore-pppfs"
parameters = {'page': 1}

# Set the column names
column_names = ['Property_details', 'Area', 'Price']

# Create an empty DataFrame
df = pd.DataFrame(columns=column_names)

# Set the number of pages 
num_pages = 20

for page in range(1, num_pages + 1):
    # Update the URL parameters
    parameters['page'] = page
    
    # Make the request
    r = requests.get(base_url, params=parameters)
    soup = BeautifulSoup(r.content, 'html.parser')
    
    # Extract the data
    lists = soup.find_all('div', class_="mb-srp__card")
    for i in lists:
        title = i.find('h2', class_="mb-srp__card--title").text
        area = i.find('div', class_="mb-srp__card__summary--value").text
        price = i.find('div', class_="mb-srp__card__price--amount").text
        Area_type = i.find('div', class_="mb-srp__card__summary--label").text
        no_of_sqfeet = i.find('div', class_="mb-srp__card__summary--value").text
        list1 = [title, area, price, Area_type, no_of_sqfeet]
        print(list1)
        row = {'Property_details': title, 'Area': area, 'Price': price, 'Area Type': Area_type, 'Number of Sq ft': no_of_sqfeet}
        # Append the row to the DataFrame
        df = df.append(row, ignore_index=True)

# Print the DataFrame
print(df)
```
